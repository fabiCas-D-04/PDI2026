ploteo por canales 

```python
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [4,3]
img=cv2.imread('leguitoskeleton.jpg', cv2.IMREAD_COLOR)
#IMAGEN ORIGINAL Y DIVIDIVA POR CANALES

plt.subplot(141)
plt.title('imagen original')
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(142)
plt.title('CANAL ROJO')
plt.imshow(img[:,:,0],cmap='gray')

plt.subplot(143)
plt.title('CANAL VERDE')
plt.imshow(img[:,:,1],cmap='gray')

plt.subplot(144)
plt.title('CANAL AZUL')
plt.imshow(img[:,:,2],cmap='gray')

plt.show()
```

con colores relleno

```python
import cv2
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 10]
img=cv2.imread('leguitoskeleton.jpg',cv2.IMREAD_GRAYSCALE)
imgRGB=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r=imgRGB[:,:,2]
g=imgRGB[:,:,1]
b=imgRGB[:,:,0]

#crear imagen por canal
r_img=np.zeros_like(imgRGB)
r_img[:,:,0]=r

g_img=np.zeros_like(imgRGB)
g_img[:,:,1]=g

b_img=np.zeros_like(imgRGB)
b_img[:,:,2]=b

plt.figure(figsize=(10,5))

plt.subplot(141)
plt.title('imagen original')
plt.imshow(imgRGB)
plt.axis('off')

plt.subplot(142)
plt.title('imagen original')
plt.imshow(r_img)
plt.axis('off')

plt.subplot(143)
plt.title('imagen original')
plt.imshow(g_img)
plt.axis('off')

plt.subplot(144)
plt.title('imagen original')
plt.imshow(b_img)
plt.axis('off')

```

generacion de falso color son cmap

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [18, 14]
img = cv2.imread('leguitoskeleton.jpg', cv2.IMREAD_COLOR)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]

plt.subplot(4,4,1)
plt.title('Imagen Original')
plt.imshow(imgRGB)
plt.axis('off')

plt.subplot(4,4,2)
plt.title('CANAL ROJO')
plt.imshow(r, cmap='Reds')
plt.axis('off')

plt.subplot(4,4,3)
plt.title('CANAL VERDE')
plt.imshow(g, cmap='Greens')
plt.axis('off')

plt.subplot(4,4,4)
plt.title('CANAL AZUL')
plt.imshow(b, cmap='Blues')
plt.axis('off')

plt.subplot(4,4,5)
plt.title('CANAL PLOMO')
plt.imshow(r, cmap='gray')
plt.axis('off')

plt.subplot(4,4,6)
plt.title('CANAL ROSA')
plt.imshow(g, cmap='pink')
plt.axis('off')

plt.subplot(4,4,7)
plt.title('CANAL MORADO')
plt.imshow(b, cmap='Purples')
plt.axis('off')

plt.subplot(4,4,8)
plt.title('CANAL AMARILLO')
plt.imshow(r, cmap='YlOrRd')
plt.axis('off')

plt.subplot(4,4,9)
plt.title('CANAL NARANJA')
plt.imshow(r, cmap='Oranges')
plt.axis('off')

plt.subplot(4,4,10)
plt.title('CANAL CIAN')
plt.imshow(b, cmap='BuGn')
plt.axis('off')

plt.subplot(4,4,11)
plt.title('CANAL MAGENTA')
plt.imshow(g, cmap='RdPu')
plt.axis('off')

plt.subplot(4,4,12)
plt.title('CANAL ESPECTRO')
plt.imshow(r, cmap='Spectral')
plt.axis('off')

plt.subplot(4,4,13)
plt.title('CANAL VIBRANTE')
plt.imshow(b, cmap='viridis')
plt.axis('off')

plt.subplot(4,4,14)
plt.title('CANAL PLASMA')
plt.imshow(r, cmap='plasma')
plt.axis('off')

plt.subplot(4,4,15)
plt.title('CANAL INFERNO')
plt.imshow(g, cmap='inferno')
plt.axis('off')

plt.subplot(4,4,16)
plt.title('CANAL TWILIGHT')
plt.imshow(b, cmap='twilight')
plt.axis('off')

plt.tight_layout()
plt.show()
```
