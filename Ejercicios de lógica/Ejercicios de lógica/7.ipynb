Desarrolla una aplicación que ordene números a través del método burbuja.

```python
try:
    cantidad = int(input("¿Cuántos números quieres ordenar?: "))
    numeros = []
    
    for i in range(cantidad):
        num = int(input(f"Introduce el número {i+1}: "))
        numeros.append(num)
    
    print(f"Números originales: {numeros}")
    
  
    n = len(numeros)
    for i in range(n):
        for j in range(0, n - i - 1):
            if numeros[j] > numeros[j + 1]:
                numeros[j], numeros[j + 1] = numeros[j + 1], numeros[j]
    
    print(f"Números ordenados: {numeros}")
except ValueError:
    print("Error: Debes introducir números enteros válidos")
```
