Desarrolle una aplicación que permita identificar si una cadena introducida por teclado es
un palíndromo ( oso es palíndromo )


```python
texto = input("Introduce una cadena de texto: ")
cadena_limpia = texto.lower().replace(" ", "")

if cadena_limpia == cadena_limpia[::-1]:
    print(f"'{texto}' ES un palíndromo")
else:
    print(f"'{texto}' NO es un palíndromo")
```
