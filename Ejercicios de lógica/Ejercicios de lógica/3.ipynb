Desarrolle una aplicación que permita ingresar diferentes números por teclado con un limite
definido inicialmente (no menor a 3 números) y determine cual de ellos es el mayor y el menor.

```python
try:
    limite = int(input("Introduce cuántos números quieres ingresar (mínimo 3): "))
    if limite < 3:
        print("El límite debe ser al menos 3 números")
    else:
        numeros = []
        for i in range(limite):
            num = int(input(f"Introduce el número {i+1}: "))
            numeros.append(num)
        
        mayor = numeros[0]
        menor = numeros[0]
        for num in numeros:
            if num > mayor:
                mayor = num
            if num < menor:
                menor = num
        
        print(f"Números ingresados: {numeros}")
        print(f"El número mayor es: {mayor}")
        print(f"El número menor es: {menor}")
except ValueError:
    print("Error: Debes introducir números enteros válidos")
```
